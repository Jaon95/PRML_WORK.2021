# Section d: Viterbi Algorithm for the Best State Sequence

```python

paramfile = 'models/partofspeech.browntags.trained'
datafile ='data/ambiguous_sents.obs'

corpus = HMM.load_observations(datafile)
model.load(paramfile)

outputfile = os.path.splitext(datafile)[0]+'.tagged.obs'
with codecs.open(outputfile, 'w', 'utf8') as o:
    for observation in corpus:
        delta,psai,max_p,i_s_l=model.viterbi(observation)
        o.write(str(observation))
```

# Section e: Forward and Backward Algorithms

```python
paramfile = 'models/partofspeech.browntags.trained'
datafile = 'data/ambiguous_sents.obs'
corpus = HMM.load_observations(datafile)
model.load(paramfile)
outputfile = os.path.splitext(datafile)[0]+'.forwardprob'
with open(outputfile, 'w') as o:
    for observation in corpus:
        # print(observation.outputseq)
        o.write(str(model.forward_probability(observation))+'\n')
```

```python
paramfile = 'models/partofspeech.browntags.trained'
datafile = 'data/ambiguous_sents.obs'
corpus = HMM.load_observations(datafile)
model.load(paramfile)
outputfile = os.path.splitext(datafile)[0]+'.backwardprob'
with open(outputfile, 'w') as o:
    for observation in corpus:
        # print(observation.outputseq)
        o.write(str(model.backward_probability(observation))+'\n')
```

# Section f: Unsupervised Learning with Baum Welch

```python
obsfile = 'data/english_words.obs'
paramfile = 'models/two_english'

model.load(paramfile)

corpus = HMM.load_observations(obsfile)
log_likelihood = model.learn_unsupervised(corpus, 0.01, False, False, 1)
#write the trained model
print ("The final model's log likelihood is", log_likelihood)
corpusbase = os.path.splitext(os.path.basename(obsfile))[0]
model.dump(paramfile+'.'+corpusbase+'.trained')
```
